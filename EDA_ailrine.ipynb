```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'User_ID': [101, 102, 103, 104, 105, 106],
    'Gender': ['M', 'F', 'M', 'F', np.nan, 'M'],
    'Age': ['18-25', '26-35', '36-45', '18-25', '55+', '46-50'],
    'City_Category': ['A', 'B', 'C', 'A', 'B', np.nan],
    'Stay_In_Current_City_Years': ['1', '2', '4+', '3', '2', '4+'],
    'Product_Category_1': [3, 1, 5, 2, 3, 5],
    'Product_Category_2': [np.nan, 6, np.nan, 8, 6, 9],
    'Product_Category_3': [15, np.nan, np.nan, 16, 15, np.nan],
    'Purchase': [12000, 15000, 18000, 20000, np.nan, 25000],
    'Date_of_Journey': ['01/03/2019', '12/05/2019', '25/12/2019', '13/01/2020', '19/06/2019', '5/8/2020'],
    'Arrival_Time': ['04:25', '22:10', '5m', '18:50', '09:15', '23:45'],
    'Dep_Time': ['02:00', '20:15', '23:00', '16:30', '06:45', '21:10'],
    'Total_Stops': ['non-stop', '1 stop', '2 stops', '1 stop', 'non-stop', np.nan],
    'Airline': ['IndiGo', 'Jet Airways', 'Air India', 'IndiGo', 'SpiceJet', 'IndiGo']
}
df=pd.DataFrame(data)
df.head()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['City_Category']=df['City_Category'].fillna(df['City_Category'].mode()[0])
df['City_Category']
```

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_2']
```

```python
df['Total_Stops']=df['Total_Stops'].fillna('1 stop')
```

```python
df=df.drop(['User_ID'],axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
scaler=LabelEncoder()
df['Gender']=scaler.fit_transform(df['Gender'])
```

```python
df['Gender']
```

```python
df['Age']=scaler.fit_transform(df['Age'])
```

```python
df['Age']
```

```python
df=pd.get_dummies(df,columns=['City_Category'])
```

```python
df
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','',regex=False).astype(int)
```

```python
df['Stay_In_Current_City_Years']
```

```python
df['Journey_Day']=pd.to_datetime(df['Date_of_Journey'],dayfirst=True).dt.day
```

```python
df['Journey_Month']=pd.to_datetime(df['Date_of_Journey'],dayfirst=True).dt.month
df['Journey_Year']=pd.to_datetime(df['Date_of_Journey'],dayfirst=True).dt.year
```

```python
df.drop(['Date_of_Journey'],axis=1)
```

```python
df=df[df['Arrival_Time']!='5m']
```

```python
df
```

```python
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0].astype(int)
```

```python
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df['Depature_Hour']=df['Dep_Time'].str.split(':').str[0].astype(int)
df['Depature_Min']=df['Dep_Time'].str.split(':').str[0].astype(int)
df=df.drop('Dep_Time',axis=1)
```

```python
df.tail()
```

```python
df['Total_Stops']=scaler.fit_transform(df['Total_Stops'])
df['Total_Stops']=df['Total_Stops'].map({
    0:1,1:0
})
```
